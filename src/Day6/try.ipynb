```python
import json

# Load the states.json file
with open(r'C:\Users\NikhilJain\py-wiz\src\Day6\states.json', 'r') as file:
    data = json.load(file)

# 1. Extract names of states with multiple area codes
states_with_multiple_area_codes = [
    state['name'] for state in data if len(state['area_codes']) > 1
]

# 2. Calculate total number of area codes for each state
total_area_codes = {
    state['name']: len(state['area_codes']) for state in data
}

# 3. Identify the state with the highest number of area codes
state_with_max_area_codes = max(total_area_codes, key=total_area_codes.get)
max_area_codes = total_area_codes[state_with_max_area_codes]

# Print the results
print("States with multiple area codes:", states_with_multiple_area_codes)
print("Total area codes per state:", total_area_codes)
print(
    f"State with the highest number of area codes: {state_with_max_area_codes} ({max_area_codes} area codes)"
)
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = r"https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv"
titanic = pd.read_csv(url)

# 1. Check how data is distributed for numerical columns
print("Numerical column distribution:")
print(titanic.describe())

# 2. Check correlation of Fare with other columns
numerical_columns = titanic.select_dtypes(include=['number'])  # Select only numerical columns
correlation = numerical_columns.corr()  # Calculate correlation
fare_correlation = correlation['Fare'].sort_values(ascending=False)  # Use 'Fare' instead of 'fare'
print("\nCorrelation of 'Fare' with other numerical columns:")
print(fare_correlation)

most_influencing = fare_correlation.index[1]  # Exclude 'Fare' itself
print(f"\nMost influencing column for Fare: {most_influencing}")



# 3. Find outliers in the ticket fare column
# 3. Find outliers in the ticket fare column
fare_q1 = titanic['Fare'].quantile(0.25)  # Corrected to 'Fare'
fare_q3 = titanic['Fare'].quantile(0.75)  # Corrected to 'Fare'
fare_iqr = fare_q3 - fare_q1

lower_bound = fare_q1 - 1.5 * fare_iqr
upper_bound = fare_q3 + 1.5 * fare_iqr
outliers = titanic[(titanic['Fare'] < lower_bound) | (titanic['Fare'] > upper_bound)]  # Corrected to 'Fare'

print(f"\nNumber of outliers in Fare column: {outliers.shape[0]}")
print(outliers['Fare'].head())  # Corrected to 'Fare'


# 4. Plot survival rate based on passenger class
sns.barplot(data=titanic, x='Pclass', y='Survived')
plt.title('Survival Rate Based on Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# 5. Analyze the survival rate based on age groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teenager', 'Adult', 'Middle-Aged', 'Senior']
titanic['age_group'] = pd.cut(titanic['Age'], bins=bins, labels=labels)

age_group_survival = titanic.groupby('age_group')['Survived'].mean()
print("\nSurvival rate based on age groups:")
print(age_group_survival)

age_group_survival.plot(kind='bar', color='skyblue')
plt.title('Survival Rate Based on Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.show()


```

```python
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the vehicle data
@st.cache
def load_data():
    # Load the CSV file, adjust path as needed
    url = r'C:\Users\NikhilJain\py-wiz\src\Day6\cars_clus.csv'
    df = pd.read_csv(url)
    return df

# Load the dataset
df = load_data()

# 1. Display data and select box for manufacturer filter
st.title('Vehicle Data Explorer')

# Create a select-box for selecting car manufacturer
manufacturers = df['manufact'].unique()
selected_manufacturer = st.selectbox('Select Car Manufacturer', manufacturers)

# Filter data based on selected manufacturer
filtered_df = df[df['manufact'] == selected_manufacturer]

# Display the filtered data
st.subheader(f"Data for {selected_manufacturer}")
st.write(filtered_df)

# 2. Plot 4 types of relevant graphs
st.subheader('Visualizations')

# Plot 1: Scatterplot of Engine Size vs Fuel Efficiency
st.write("### Engine Size vs Fuel Efficiency (Scatter Plot)")
fig, ax = plt.subplots()
sns.scatterplot(x=filtered_df['engine_s'], y=filtered_df['mpg'], ax=ax)
ax.set_title('Engine Size vs Fuel Efficiency')
ax.set_xlabel('Engine Size (L)')
ax.set_ylabel('Fuel Efficiency (mpg)')
st.pyplot(fig)

# Plot 2: Distribution of Vehicle Prices (Histogram)
st.write("### Distribution of Vehicle Prices (Histogram)")
fig, ax = plt.subplots()
sns.histplot(filtered_df['price'], kde=True, ax=ax)
ax.set_title('Distribution of Vehicle Prices')
ax.set_xlabel('Price')
ax.set_ylabel('Count')
st.pyplot(fig)

# Plot 3: Boxplot of Engine Size by Vehicle Type
st.write("### Engine Size by Vehicle Type (Box Plot)")
fig, ax = plt.subplots()
sns.boxplot(x=filtered_df['type'], y=filtered_df['engine_s'] ,ax=ax)
ax.set_title('Engine Size by Vehicle Type')
ax.set_xlabel('Vehicle Type')
ax.set_ylabel('Engine Size (L)')
st.pyplot(fig)

# Plot 4: Price vs Weight (Scatter Plot)
st.write("### Price vs Weight (Scatter Plot)")
fig, ax = plt.subplots()
sns.scatterplot(x=filtered_df['curb_wgt'], y=filtered_df['price'], ax=ax)
ax.set_title('Price vs Weight')
ax.set_xlabel('Weight (lbs)')
ax.set_ylabel('Price')
st.pyplot(fig)

```
