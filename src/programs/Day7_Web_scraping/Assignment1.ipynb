```python
driver = webdriver.Chrome()

# URL to scrape
url = "https://www.accuweather.com/en/in/mumbai/204842/weather-forecast/204842"

# Open the URL
driver.get(url)
```

1.Using xpath of specific element



```python


def find_day(lock,xpath1):
    driver.get(url)
    driver.implicitly_wait(10)
    data=driver.find_element(By.XPATH ,xpath1)
    print(f"Todaty is {data.text}")

def find_temp_max(lock,xpath2):
    driver.get(url)
    driver.implicitly_wait(10)
    data=driver.find_element(By.XPATH ,xpath2)
    print(f"Highest temparature is {data.text}")

def find_temp_min(lock,xpath3):
    driver.get(url)
    driver.implicitly_wait(10)
    data=driver.find_element(By.XPATH ,xpath3)
    print(f"Lowest temprature {data.text}")


```

```python
lock=threading.Lock()

xpathone="/html/body/div/div[7]/div[1]/div[1]/div[1]/a/div[1]/p"
xpath2="/html/body/div/div[7]/div[1]/div[1]/div[1]/a/div[2]/div[1]/p/b"
xpath3="/html/body/div/div[7]/div[1]/div[1]/div[1]/a/div[2]/div[2]/p/b"

th1=threading.Thread(target=find_day,args=(lock,xpathone))
th2=threading.Thread(target=find_temp_max,args=(lock,xpath2))
th3=threading.Thread(target=find_temp_min,args=(lock,xpath3))

th1.start()
th2.start()
th3.start()

th1.join()
th2.join()
th3.join()

driver.quit()
```

2 using class 

```python
time.sleep(10)
page_source = driver.page_source
page_source
driver.quit()
```

```python
# Get the page source and parse it with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# Extract weather information
forecast_items = soup.find_all('a', class_='daily-list-item')
forecast_items
```

```python
# Iterate through each forecast_item
for item in forecast_items:
    # Etract day
    day = item.find('p', class_='day').text

    # Extract high and low temperatures 
    high_temp = item.find('span', class_='temp-hi').text
    low_temp = item.find('span', class_='temp-lo').text

    # Extract the weather phrase
    weather_phrase = item.find('p', class_='no-wrap').text


    # Print the extracted details
    print(f"Day: {day}")
    print(f"High Temperature: {high_temp}")
    print(f"Low Temperature: {low_temp}")
    print(f"Weather: {weather_phrase}")
    print("-" * 40)
```

```python
import random
import threading
import time
```

```python
def place_order(customer):
    coffiee_type=random.choice(["Espresso", "Latte", "Cappuccino", "Americano", "Mocha"])
    size=random.choice(['Small','Medium','Large'])
    milk_type=random.choice(['milk','no milk'])

    print(f"Customer {customer} is ordering: {size} {coffiee_type} with {milk_type}")
    
    prep_time=random.randint(10,15)
    print(f"Preparing order for Customer {customer}... (Estimated {prep_time} sec)")
    time.sleep(prep_time)

    
    print("**************************************************************************************")
    print(f"\nOrder ready for Customer {customer}: {size} {coffiee_type} with {milk_type} ")

def main():
    print("**************************************************************************************")
    print("Welcome to the Coffee Shop! \n")
    print("**************************************************************************************")
    customer=5
    threads=[]  


    for i in range(customer):
        thread=threading.Thread(target=place_order,args=(i+1,))
        threads.append(thread)
        thread.start()

    for thre in threads:
        thre.join()

    print("All orders have been processed. Enjoy your coffee! ")
```

```python

main()
```

```python
import requests
import time
import multiprocessing

# URL 
url = "https://picsum.photos/2000/3000"
```

```python
# Function to download an image
def download_image(index):
    print("here")
    response = requests.get(url)
    print(response.content)
    with open(f"image_{index}.jpg", 'wb') as f:
        f.write(response.content)
    print(f"Downloaded image_{index}.jpg")  
```

```python
# Without multiprocessing
def download_images_sequential():
    start_time = time.time()
    for i in range(5):
        download_image(i)
    end_time = time.time()
    print(f"Sequential download time: {end_time - start_time:.2f} seconds")
```

```python
download_images_sequential()
```

```python
# With multiprocessing
def download_images_multiprocessing():
    start_time = time.time()    
    processes = []
    for i in range(5):
        process = multiprocessing.Process(target=download_image, args=(i,))
        print("here")
        processes.append(process)
        process.start()
    
    for process in processes:
        print("in join")
        process.join()

    end_time = time.time()
    print(f"Multiprocessing download time: {end_time - start_time:.2f} seconds")
```

```python
download_images_multiprocessing()
```
