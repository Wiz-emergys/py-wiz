```python
def calc_square(lock,lst):
    print("Calculating Squares: ")
    lock.acquire()
    for ele in lst:
        time.sleep(0.1)
        # cpu will be idle for 0.2 seconds
        print(f"Square of {ele}: {ele ** 2}")
        
    lock.release()
       
 
 
def calc_cube(lock,lst):
    print("Calculating Cubes: ")
    lock.acquire()
    for ele in lst:
        time.sleep(0.1)
        
        print(f"Cube of {ele}: {ele ** 3}")
    lock.release()
        
```

```python
import threading
import time
```

```python
list_of_num=[1,2,4,5,6]

lock=threading.Lock()
thread1=threading.Thread(target=calc_square,args=(lock,list_of_num))

thread2=threading.Thread(target=calc_cube,args=(lock,list_of_num))

thread1.start()
thread2.start()


thread1.join()
print("thread one completed")
thread2.join()
print("thead 2 completed")



```

Synchoronization

The synchornization is basically a concept used when you are using shared resources like shared memory or msg passing 
the basic idea is to keep other users locked till one user completes its exectution or task and allow others based on priority this helps in deadlock prevention and deadlock avoidence .the concept is quite simple and efficient when working with threds and paralle processing is used
